I have cleaned the dataset by filling the null spaces with the average of the existing values min each respective column. This approach helps maintain the overall data distribution and prevents loss of information that might occur if rows with missing values were removed.

```python
import seaborn as sns
import matplotlib.pyplot as plt
top10_countries = db.groupby('country')['Estimated number of incident cases (all forms)'].sum().sort_values(ascending=False).head(10).index
    
top10_data = db[db['country'].isin(top10_countries)]

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=top10_data,
    x='year',
    y='Estimated number of incident cases (all forms)',
    hue='country'
)
plt.title('Top 10 Countries by Estimated TB Incident Cases')
plt.xlabel('Year')
plt.ylabel('Estimated number of incident cases (all forms)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

The line chart illustrates the estimated number of TB incidents in the top ten countries over the years. Each line represents a country, showing how the TB burden has changed annually. This visualization helps identify trends, highlight countries with the highest incidence, and observe whether TB cases are increasing, stable, or decreasing over time.


```python
import seaborn as sns
import matplotlib.pyplot as plt
top10_countries = db.groupby('country')['Estimated HIV in incident TB (percent)'].sum().sort_values(ascending=False).head(10).index
top10_data = db[db['country'].isin(top10_countries)]
plt.figure(figsize=(12, 6))
bar_chart=sns.barplot(y='Estimated HIV in incident TB (percent)',x='country',data=top10_data,hue='year')

```

The chart illustrates the trend of HIV in incident TB (%) across the top ten countries, highlighting a reduction in the rate of HIV-positive TB cases over time.

```python
top10_data = (
    db.groupby('country')["Estimated incidence of TB cases who are HIV-positive per 100 000 population"]
    .sum()
    .sort_values(ascending=False)
    .head(10))
    
plt.figure(figsize=(12, 6))
plt.pie(
    top10_data.values,
    labels=top10_data.index,autopct='%1.1f%%')
```

The pie chart shows the distribution of TB cases who are HIV-positive (per 100,000 population) among the top ten countries, highlighting which countries bear the highest burden of HIV-associated TB.

```python
top10_data = (
    db.groupby('country')["Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population"]
    .sum()
    .sort_values(ascending=False)
    .head(10))
plt.figure(figsize=(12, 6))
plt.pie(
    top10_data.values,
    labels=top10_data.index,autopct='%1.1f%%',
    startangle=140,
    )
```

The pie chart illustrates the distribution of TB mortality (all forms, excluding HIV) per 100,000 population across the top ten countries, highlighting which countries experience the highest TB death rates.

```python
top10_countries = (
    db.groupby('country')["Estimated number of deaths from TB (all forms, excluding HIV)"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index)
top10_data = db[db['country'].isin(top10_countries)]
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top10_data,
    x='year',
    y='Estimated number of deaths from TB (all forms, excluding HIV)',
    hue='country'
)
```

The bar chart presents the estimated number of deaths from TB (all forms, excluding HIV) in the top ten countries. Each bar represents a country, allowing for a clear comparison of TB mortality across nations. This visualization highlights the countries with the highest TB death burden, providing insight into where TB remains a significant public health challenge.

```python
sns.scatterplot(data=db,x='year',y="Estimated number of deaths from TB in people who are HIV-positive")
```

The scatter plot depicts the estimated number of TB deaths among HIV-positive individuals over the years, illustrating the trend in HIV-associated TB mortality over time.

```python
top10_countries = (
    db.groupby('country')["Estimated TB case fatality ratio"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index)
top10_data = db[db['country'].isin(top10_countries)]
plt.figure(figsize=(18, 4))
sns.boxplot(
    data=top10_data,
    x='country',
    y='Estimated TB case fatality ratio',
    palette='Set2')
```

The box plot illustrates the distribution of estimated TB case fatality ratios in the top ten countries with the highest TB burden. Each box represents a country, showing the median, range, and variation of TB case fatality ratios. This visualization highlights differences in TB mortality risk among the highest-burden countries and allows for comparison of how fatal TB cases are across nations.

```python
country_data = db.groupby('country')[[
    'Estimated incidence (all forms) per 100 000 population',
    'Estimated mortality of TB cases (all forms) per 100 000 population',
    'Estimated TB case fatality ratio'
]].mean()

plt.figure(figsize=(12, 8))
sns.heatmap(country_data.sort_values('Estimated incidence (all forms) per 100 000 population', ascending=False).head(20),
            cmap='Reds', annot=True, fmt=".1f")
plt.title('Top 20 Countries by TB Incidence, Mortality, and Fatality Ratio')
plt.show()
```

The heatmap displays key TB indicators‚Äîestimated incidence per 100,000 population, estimated mortality per 100,000 population, and TB case fatality ratio‚Äîacross the top 20 countries with the highest TB burden. Each cell represents a country-indicator pair, with color intensity reflecting the magnitude. This visualization allows for quick comparison of TB burden, mortality, and case fatality risk across nations, highlighting countries with the highest impact.

```python
yearly_tb = (
    db.groupby('year')["Estimated incidence (all forms) per 100 000 population"]
    .mean().reset_index()  )
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=yearly_tb,
    x='year',
    y='Estimated incidence (all forms) per 100 000 population',
    marker='o'
)
```

The line chart illustrates the trend of estimated TB incidence (all forms) per 100,000 population over the years. Each point represents the average incidence for all countries, showing whether TB incidence has increased, remained stable, or decreased globally.

```python
plt.figure(figsize=(12,6))
sns.barplot(data=db, x='year', y='Case detection rate (all forms) [also known as TB treatment coverage], percent', palette='Blues_d')
plt.title('Global TB case detection rate Over Years')
plt.xlabel('Year')
plt.ylabel('Estimated Number of Incident Cases')
plt.show()
```

The bar chart shows the global TB case detection rate (also known as TB treatment coverage) over the years. Each bar represents the average percentage of TB cases detected per year, illustrating progress in identifying and treating TB worldwide.

General Insights: TB Data Overview

The visualizations show that global TB incidence and mortality are gradually declining, reflecting the impact of prevention, treatment, and awareness programs. High-burden countries still carry the majority of TB cases and deaths, highlighting ongoing challenges.

HIV-associated TB is decreasing in the top affected countries, indicating improvements in integrated TB-HIV management. Box plots of case fatality ratios reveal variation across countries, reflecting differences in healthcare access, treatment quality, and adherence.

Case detection rates are improving over time but remain below optimal levels, suggesting underdiagnosis and surveillance gaps. Heatmaps of the top 20 countries show which nations have the highest burden in terms of incidence, mortality, and fatality, helping to prioritize targeted interventions.

Overall, the data indicate global progress in TB control, while emphasizing the need for continued focus on high-burden countries, improving case detection, and reducing mortality.
